```python
import json
from io import StringIO
from datetime import datetime, timedelta

import numpy as np
import pandas as pd


NB_ROWS = 500

channels = ["Téléphone", "Email", "Chat", "Réseaux Sociaux"]
languages = ["FR", "EN", "ES", "DE", "IT", "NL", "PT"]
segments = ["Famille", "Abonné Annuel", "Primo-visiteur", "VIP"]


def make_fake_data(nb=NB_ROWS):
    np.random.seed(42)

    data = []
    start = datetime(2026, 5, 21, 8, 0)

    for idx in range(nb):

        channel = np.random.choice(channels)
        lang = np.random.choice(languages)
        segment = np.random.choice(segments)

        # valeurs un peu différentes selon le canal
        if channel == "Email":
            duration = round(np.random.normal(30, 7), 2)
            score = np.random.choice([1, 2, 3, 4, 5], p=[0.1, 0.1, 0.2, 0.4, 0.2])

        elif channel == "Chat":
            duration = round(np.random.normal(12, 3), 2)
            score = np.random.choice([1, 2, 3, 4, 5], p=[0.05, 0.05, 0.15, 0.45, 0.3])

        else:
            duration = round(np.random.normal(8, 2.5), 2)
            score = np.random.choice([1, 2, 3, 4, 5], p=[0.05, 0.1, 0.2, 0.4, 0.25])

        # éviter les valeurs négatives
        if duration < 1:
            duration = 1

        date_value = start + timedelta(
            seconds=np.random.randint(0, 36000)
        )

        # ancien export vs nouveau format
        if idx % 2 == 0:
            opened = date_value.strftime("%Y-%m-%d %H:%M:%S")
        else:
            opened = f"{date_value.strftime('%d/%m/%Y')} {date_value.hour}h{date_value.minute:02d}"

        row = {
            "interaction_id": f"DIS-{1000 + idx}",
            "agent_id": f"AG-{np.random.randint(1, 501):03d}",
            "channel": channel,
            "language": lang,
            "customer_segment": segment,
            "opened_at": opened,
            "duration_minutes": duration,
            "csat_score": int(score),
        }

        data.append(row)

    return json.dumps(data)


def export_metrics(raw_json, keep_langs=None, filename="crc_metrics.csv"):

    if keep_langs is None:
        keep_langs = languages

    df = pd.read_json(StringIO(raw_json))

    # certains fichiers ont "14h32"
    df["opened_at"] = df["opened_at"].astype(str).str.replace("h", ":")

    df["opened_at"] = pd.to_datetime(
        df["opened_at"],
        format="mixed"
    )

    df = df[df["language"].isin(keep_langs)]

    print("rows gardées :", len(df))

    print("\nAHT global")
    print(round(df["duration_minutes"].mean(), 2), "min")

    print("\nAHT par canal")
    print(
        df.groupby("channel")["duration_minutes"]
        .mean()
        .round(2)
    )

    print("\nCSAT moyen par segment")
    print(
        df.groupby("customer_segment")["csat_score"]
        .mean()
        .round(2)
    )

    df.to_csv(filename, index=False, encoding="utf-8")

    print("\nfichier exporté :", filename)

    return df


if __name__ == "__main__":

    fake_json = make_fake_data()

    export_metrics(
        fake_json,
        keep_langs=["FR", "EN"]
    )
```
