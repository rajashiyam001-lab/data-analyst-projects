```python
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
df = df.rename(columns={
    'customer_id': 'id_client',
    'âge': 'age',
    'gender': 'sexe',
    'item_purchased': 'article_achete',
    'category': 'categorie',
    'purchase_amount': 'montant_achat',
    'location': 'localisation',
    'size': 'taille',
    'color': 'couleur',
    'season': 'saison',
    'review_rating': 'note_avis',
    'subscription_status': 'statut_abonnement',
    'shipping_type': 'type_livraison',
    'discount_applied': 'remise_appliquee',
    'promo_code_used': 'code_promo_utilise',
    'previous_purchases': 'achats_precedents',
    'payment_method': 'mode_paiement',
    'frequency_of_purchases': 'frequence_achats'
})
```

```python
df.columns
```

```python
labels = ['jeune-adulte', 'adulte', 'age-moyen','personne-agee']
df['groupe_age'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','groupe_age']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-weekly': 14,
    'Annually': 365,
    'Every 3 months': 90
}
df['frequence_achats_jours'] = df['fréquence_achats'].map(frequency_mapping)


```

```python
df[['frequence_achats_jours','fréquence_achats']].head(10)
```

```python
df['frequence_achats_jours'] = df['frequence_achats_jours'].fillna(0).astype(int)

```

```python
df[['frequence_achats_jours','fréquence_achats']].head(10)
```
